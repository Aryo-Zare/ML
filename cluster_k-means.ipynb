```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans as km
```

```python
cc = np.array([ [1,1] , [4,2] , [2,3] , [3 , 0.5] ])

# it would have been better to name them  scat_lb for scatter_label
# x is a 2 dimensional array : x-y coordinate of each scatter dot.
# y is the label of each dot (belonging to which cluster).
x , y = make_blobs(
	n_samples=4000,
	n_features=2,
	centers=cc,
	cluster_std=0.3,
	# center_box = (-5 , 5),
	random_state=0
)


# 2 features : 1 in x axis & 1 in y axis.
plt.scatter(x[: , 0] , x[: ,1])
```

```python
model = km(
	n_clusters=4,  # this can be changed depending on your preferences.
	init = 'k-means++' ,
	n_init='auto'
)

model.fit(x)
```

```python
# cts : centers.
# yields a numpy array containing the cluster centers.
cts = model.cluster_centers_

# lbl : labels : label of each point : belonging to which cluster.
lbl = model.labels_

```

```python
# 0,1,2,3 : the label of each cluster.
labels = [0 , 1 , 2 , 3]
colors = [ 'b' , 'r' , 'g' , 'c' ]


for i in labels :
	mask = (lbl==i) # a boolean array.
	plt.scatter( x[mask,0] , x[mask,1] , c=colors[i] )

plt.scatter(cts[:,0] , cts[:,1] , s=150 , c='k')
```

```python

# you want to select the blue cluster (selecting all items in it) :
# according to the plot, its label is '0'
# it has a shape of about (1000 , 2).
mask = (lbl==0)
blue_cluster = x[ mask , : ]

# only plotting the blue_cluster :
plt.scatter(blue_cluster[:,0] , blue_cluster[:,1])
```
